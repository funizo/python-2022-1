# 처음시작하는 코딩 파이선 기말고사 과제
`numpy`와 `matplotlib` 을 이용한 직군별 연차 연봉을 알아보는 프로그램입니다.

```python
#모듈 설치
import numpy as np
import matplotlib.pyplot as plt
import turtle
import nest_asyncio
nest_asyncio.apply()
```

```python
#직군별 연차 연봉
front = np.array([3353,3503,3761,4005,4395,4599],dtype=int) #프론트엔드개발자
end = np.array([3489,3718,3936,4286,4602,4817],dtype=int) #백앤드개발자
ios = np.array([3530,3766,3900,4163,4351,4539],dtype=int) #ios앱개발자
android = np.array([3306,3575,3736,4038,4358,4477],dtype=int) #안드로이드앱개발자
python = np.array([3481,3816,4078,4369,4545,5021],dtype=int) #파이선개발자
bigdata = np.array([3546,3850,4366,4666,5078,5323],dtype=int) #빅데이터개발자



frontsum = np.sum(front,axis=0)
frontavg = round(frontsum/6)
print("프론트엔드 개발자 평균 연봉",frontavg,"만원")

endsum = np.sum(end,axis=0)
endavg = round(endsum/6)
print("백앤드 개발자 평균 연봉",endavg,"만원")

iossum = np.sum(ios,axis=0)
iosavg = round(iossum/6)
print("ios개발자 평균 연봉",iosavg,"만원")

androidsum = np.sum(android,axis=0)
androidavg = round(androidsum/6)
print("안드로이드 개발자 평균 연봉",androidavg,"만원")

pythonsum = np.sum(python,axis=0)
pythonavg = round(pythonsum/6)
print("파이선 개발자 평균 연봉",pythonavg,"만원")

bigdatasum = np.sum(bigdata,axis=0)
bigdataavg = round(bigdatasum/6)
print("빅데이터 개발자 평균 연봉",bigdataavg,"만원")


```

```python
#차트 생성
plt.plot(front,color='red',label='front')
plt.plot(end,color='blue',label='back-end')
plt.plot(ios,color='green',label='ios')
plt.plot(android,color='orange',label='android')
plt.plot(python,color='violet',label='python')
plt.plot(bigdata,color='yellow',label='bigdata')
plt.xlabel('years')
plt.ylabel('pay')
plt.legend(loc='lower right')
```

```python
#평균연봉차트

x = np.arange(6)
developer = ['front', 'back-end', 'ios','android','python','bigdata']
values = [frontavg,endavg,iosavg,androidavg,pythonavg,bigdataavg]
colors = ['r', 'b', 'g','orange','violet','y']
plt.plot(values, 'ko--')


plt.bar(x, values,width=0.4,color=colors)
plt.xticks(x, developer)

plt.show()
```

```python
#6개 직군의 평균 연봉
total = (frontavg+endavg+iosavg+androidavg+pythonavg+bigdataavg)/6

other = np.array([3926,3635,3316,3315,3784,3582],dtype=int)  #경영,무역,서비스,생산,마케팅,바이오
othersum = np.sum(other,axis=0)
otheravg = round(othersum/6)

z = np.arange(2)
dif = ['developer','otherjob']
values = [total,otheravg]
colors = ['r','b']
plt.bar(z, values,width=0.4,color=colors)
plt.xticks(z, dif)
plt.show()

```

```python
paydif = round(total-otheravg)
print("개발자와 다른직군의 평균연봉차이는",paydif,"만원입니다")
print("우리모두 열심히 코딩하여서 높은 연봉을 받는 개발자가 됩시다!")
```
